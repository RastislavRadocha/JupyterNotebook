```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
pokemons = pd.read_csv('Pokemon.csv', index_col='#')
pokemons
```

```python
poke_groups = pokemons.groupby(by='Type 1')
max_attack = poke_groups['Attack'].max()
max_attack
```

```python
max_attack.plot(kind='barh', color=('skyblue', 'crimson'))
```

```python
plt.style.available
```

```python
plt.style.use('seaborn-v0_8-deep')
max_attack.plot(kind='barh')
```

```python
def all_designs():
    for style in plt.style.available:
        plt.style.use(style)
        max_attack.plot(kind='barh', figsize=(4,3), fontsize=6)
        print(style)
        plt.show()

all_designs()
```
