```python
cols = ['date', 'high', 'low', 'open', 'close','volume']
apple_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks
```

```python
year_2016 = apple_stocks.loc['2016-01':'2016-3']
year_2016
```

```python
mpf.plot(year_2016, type='candle',
       style='yahoo',
       title='Apple Stocks',
       ylabel='Price $',
       figratio=(3, 1),
       volume=True)
```
