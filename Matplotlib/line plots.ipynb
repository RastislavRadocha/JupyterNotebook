```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
cols = ['date', 'high', 'low', 'open', 'close',]
apple_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks
```

```python
apple_stocks.plot()
```

```python
apple_stocks.loc['2016-05'].plot()
```

```python
apple_stocks.loc['2016-05'].plot()
```

```python
apple_stocks.plot(subplots=True, figsize=(10, 5))
```

```python
apple_stocks.plot()
plt.title('Apple stocks')
plt.legend(loc='best')
plt.xlabel('Years')
plt.ylabel('Price')
plt.grid()
```
