```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
richest = pd.read_csv('TopRichestInWorld.csv')
richest
```

```python
industries = richest['Industry'].value_counts()
industries
```

```python
industries.index
```

```python
plt.pie(x=industries.values, labels=industries.index)
plt.show
```

```python
labels = industries.index[0:10]
values = industries.values[0:10]

plt.pie(x=values, labels=labels, autopct='%.2f%%')
plt.show()
```

```python
plt.figure()
plt.pie(x=values, labels=labels, colors=('skyblue', 'darkblue', 'blue'))
plt.show()
```
