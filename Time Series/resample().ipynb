```python
import pandas as pd
```

```python
cols=['date', 'high', 'low', 'open', 'close']
apple_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks.head(7)
```

```python
months = apple_stocks.resample('MS')
months
```

```python
months.groups
```

```python
months.get_group('2020-03-01 00:00:00+0000')
```

```python
months.mean().round(2)
```

```python
apple_stocks.index = apple_stocks.index.to_period('M')
months2 = apple_stocks.groupby(apple_stocks.index)
months2
```

```python
months2.groups
```

```python
months2.mean().round(2)
```

```python
months2.agg({'high':['max', 'min'], 'low':['max', 'min']}).round(2)
```
