```python
import pandas as pd
```

```python
cols=['date', 'high', 'low', 'open', 'close']
apple_stocks = pd.read_csv('AAPL.csv', usecols=cols, index_col='date', parse_dates=['date'])
apple_stocks.head(7)
```

```python
apple_stocks.index.day_name().value_counts()
```

```python
max_date = apple_stocks.index.max()
max_date
```

```python
min_date = apple_stocks.index.min()
min_date
```

```python
dates_incl_weekends = pd.date_range(start=min_date, end=max_date, freq='d')
dates_incl_weekends
```

```python
dates_incl_weekends.day_name().value_counts()
```

```python
apple_stocks.reindex(dates_incl_weekends).head(7)
```

```python
apple_stocks.reindex(dates_incl_weekends).head(7).fillna('CLOSED')
```

```python
apple_stocks.reindex(dates_incl_weekends, method='ffill').head(7)
```

```python
apple_stocks.reindex(dates_incl_weekends, method='bfill').head(7)
```

```python
apple_stocks.reindex(dates_incl_weekends, method='nearest').head(7)
```

```python
as2 = apple_stocks.reindex(dates_incl_weekends, method='ffill')
as2['day'] = as2.index.day_name()
as2.head(7)
```
