```python
import pandas as pd
```

```python
exams = pd.read_csv('exams.csv')
exams
```

```python
exams['total'] = exams['math score'] + exams['writing score'] + exams['reading score']
exams
```

```python
groups = exams.groupby(['gender', 'lunch'])
groups
```

```python
stats = groups.agg({'total': ['max', 'min', 'mean']}).round(2)
stats
```

```python
stats.sort_values(by='total', ascending=False)
```
