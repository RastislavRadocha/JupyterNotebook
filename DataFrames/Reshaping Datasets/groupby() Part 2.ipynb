```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
poke_groups = pokemon.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups.groups
```

```python
poke_max = poke_groups.max()
poke_max
```
