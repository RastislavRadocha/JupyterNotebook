```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
pokemon.groupby('Type 1').agg({'Total': ['max', 'min', 'mean']}).round(2)
```

```python
pokemon.groupby(by=['Type 1', 'Type 2']).agg({'Total': ['max', 'min', 'mean']}).round(2)
```

```python
poke_groups = pokemon.groupby(by=['Type 1', 'Type 2'], dropna=False)
poke_groups
```

```python
poke_groups.agg({'Total': ['max', 'min', 'mean'], 'Attack': 'max'}).round(2)
```
