```python
import pandas as pd
```

```python
people = pd.read_csv('people2.csv', sep=';')
people
```

```python
formatted = people['Salary'].str.replace('€', '').str.replace(',', '').astype(int) / 100
people['Salary'] = formatted
people
```

```python
people.pivot_table(values='Salary', # which values we want to use
                   columns='Name', # which columns to perform aggfunc on
                   aggfunc='sum') # which operations to perform ('mean' is default)
```

```python
people.pivot_table(values='Salary', columns='Day').round(2)
```

```python
people.pivot_table(index='Day', 
                   values='Salary', # which values we want to use 
                   columns='Gender', # which columns to perform aggfunc on
                   aggfunc='max') # which operation we want to use ('mean' is default)
```
