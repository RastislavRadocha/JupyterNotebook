```python
import pandas as pd
```

```python
medals = pd.read_csv('olympics.csv', sep=';')
medals
```

```python
medals.melt(id_vars='Country', # This is the column(s) that uniquely identify each row, and are not meant to be reshaped or transformed
            value_vars=['2021', '2022', '2023',  '2024'] # Variables which we are 'melting', thus reshaping the dataset
           )
```

```python
medals.melt(id_vars='Country',
            var_name='Year',
            value_name='Medals',
            value_vars=['2021', '2022'])
```
