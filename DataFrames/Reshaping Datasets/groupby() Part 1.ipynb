```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
pokemon['Type 1'].value_counts()
```

```python
poke_groups = pokemon.groupby(by='Type 1')
poke_groups
```

```python
poke_groups.groups
```

```python
poke_groups.get_group('Grass').head(5)
```

```python
pokemon[pokemon['Type 1'] == 'Grass'].head()
```

Why create a group and not a mask?
    - Groups are more flexible and promote code reusability
    - Prevents us from creating a million masks

```python
poke_groups.get_group('Fire').head()
```

```python
pokemon.pivot_table(index='Type 1', aggfunc='mean').round(2)
```

```python
poke_groups.mean(numeric_only=True).round(2)
```

```python
poke_groups.get_group('Grass').mean(numeric_only=True).round(2)
```
