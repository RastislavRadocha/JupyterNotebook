```python
import pandas as pd
```

```python
titanic = pd.read_excel('titanic3.xls', index_col='name')
titanic
```

```python
titanic.drop(columns=['home.dest', 'embarked', 'sibsp', 'parch'], inplace=True)
titanic
```

```python
survived_mask = titanic['survived'] == 1
titanic = titanic[survived_mask]

titanic
```

```python
titanic.drop(columns=['body', 'survived'], inplace=True)
titanic
```

```python
third_class_mask = titanic['pclass'] == 3
titanic.loc[third_class_mask, 'fare'].max()
```

```python
titanic.index.name = 'Passengers'
titanic
```

```python
free_coffee_mask = titanic['pclass'].isin([1,2])
titanic['Free Coffee'] = 0
titanic.loc[free_coffee_mask, 'Free Coffee'] = 1

titanic
```

```python
titanic.loc[:, ['pclass', 'Free Coffee']].value_counts()
```
