```python
cols=['Series_Title','IMDB_Rating', 'No_of_Votes', 'Gross', 'Released_Year', 'Genre']
```

```python
movies=pd.read_csv('imdb_top_1000.csv', usecols=cols)
movies
```

```python
movies.info()
```

```python
movies['Gross'] = movies['Gross'].fillna(0).astype('int64')
movies
```
