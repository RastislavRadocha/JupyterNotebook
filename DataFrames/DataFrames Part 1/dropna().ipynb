```python
import pandas as pd
```

```python
df = pd.read_csv('simple.csv', sep=';', index_col=['Unnamed: 0'])
df
```

```python
df.drop(columns='e', inplace=True)
df = df.head(10)
df
```

```python
df.dropna() # drops every row that contains NaN value
```

```python
df.dropna(axis=0) # removes all the rows that contain NaN, axis=0 is by default
```

```python
df.dropna(axis=1) # removes all columns that contains NaN
```

```python
df.dropna(axis=0, how='all')
```
