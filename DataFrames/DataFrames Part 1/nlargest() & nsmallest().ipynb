```python
import pandas as pd
```

```python
cols = ['Series_Title', 'IMDB_Rating', 'No_of_Votes']
```

```python
movies = pd.read_csv('imdb_top_1000.csv', usecols=cols)
movies
```

```python
movies.nlargest(n=10, columns=['No_of_Votes'])
```

```python
movies.nlargest(n=10, columns=['No_of_Votes','IMDB_Rating'])
```

```python
movies.nsmallest(n=10, columns=['No_of_Votes', 'IMDB_Rating'])
```

- ``first`` : take the first occurrence.
- ``last`` : take the last occurrence.
- ``all`` : keep all the ties of the largest item even if it means
	       selecting more than ``n`` items.

```python
movies.nlargest(n=10, columns=['IMDB_Rating'], keep='first')
```

```python
movies.nsmallest(n=10, columns=['IMDB_Rating'], keep='last')
```
