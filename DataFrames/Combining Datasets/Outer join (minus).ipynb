```python
import pandas as pd
```

```python
cols = ['Name', 'Salary']
```

```python
company21 = pd.read_csv('company21.csv', sep=';', usecols=cols)
company21
```

```python
company22 = pd.read_csv('company22.csv', sep = ';', usecols=cols)
company22
```

```python
merged_outer = pd.merge(left=company21, right=company22, 
         how='outer', 
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)

merged_outer[merged_outer['_merge'] != 'both']
```

```python
pd.merge(left=company21, right=company22, 
         how='outer', 
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```
