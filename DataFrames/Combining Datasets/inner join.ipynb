```python
import pandas as pd
```

```python
cols = ['Name', 'Salary']
```

```python
company21 = pd.read_csv('company21.csv', sep=';', usecols=cols)
company21
```

```python
company22 = pd.read_csv('company22.csv', sep = ';', usecols=cols)
company22
```

```python
pd.merge(left=company21, right=company22, 
         how='inner', # inner is the default value 
         on='Name',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
pd.merge(left=company21, right=company22, 
         how='inner', # inner is the default value 
         on='Salary',
         suffixes=('_21', '_22'),
         indicator=True)
```

```python
pd.merge(left=company21, right=company22, 
         how='inner', # inner is the default value 
         suffixes=('_21', '_22'),
         indicator=True)
```
