```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
pokemon.agg({'Attack': 'mean', 'Defense': 'max'})
```

```python
pokemon.agg({'Attack': ['min', 'max', 'std'], 'Defense': ['min', 'max', 'mean']}) # 'std' -> standard deviation
```

```python
pokemon.describe()
```

```python
pokemon.agg('max', numeric_only=True)
```

```python
pokemon_minmax= pokemon.drop(columns=['Name', 'Type 1', 'Type 2', 'Legendary'])
pokemon_minmax.agg(['max', 'min', 'mean', 'std'], numeric_only=True).round()
```
