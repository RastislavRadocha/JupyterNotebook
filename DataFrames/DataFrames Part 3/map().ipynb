```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
pokemon['Legendary'] = pokemon['Legendary'].map({False:'No', True:'Yes'})
pokemon
```

```python
pokemon['Generation'] = pokemon['Generation'].map({1:'First'})
pokemon
```

```python
def uppercase(var: str) -> str:
    return var.upper()
```

```python
pokemon = pokemon.map(str)
pokemon.map(uppercase)
```
