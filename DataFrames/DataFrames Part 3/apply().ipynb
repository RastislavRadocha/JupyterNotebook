```python
import pandas as pd
```

```python
pokemon = pd.read_csv('Pokemon.csv', index_col='#')
pokemon
```

```python
pokemon['name_length'] = pokemon['Name'].apply(len)
pokemon
```

```python
del pokemon['name_length']
```

```python
def strength(attack:int) -> str:
    """
    Categorizes an attack value into strength levels.

    Parameters:
    attack (int): The attack value to classify.

    Returns:
    str: A string indicating the strength level:
         - 'Strong' if attack > 150
         - 'Average' if 100 < attack <= 150
         - 'Weak' if attack <= 100
    """
    if attack > 150:
        return 'Strong'
    elif attack > 100:
        return 'Average'
    else:
        return 'Weak'
```

```python
pokemon['Strength'] = pokemon['Attack'].apply(strength)
pokemon
```

```python
pokemon['Strength'].value_counts()
```

```python
pokemon.value_counts('Strength')
```

```python
del pokemon['Strength']
```

```python
def score(row):
    """
    Calculates the total score of a row based on Attack, Speed, and Defense values.

    Parameters:
    row (pd.Series): A row from a pandas DataFrame, typically passed by the .apply() method when `axis=1` is used.

    Returns:
    int: The total score, which is the sum of Attack, Speed, and Defense values.
    """
    # accessing the 'Attack' value from the row
    attack: int = row['Attack']
    # accessing the 'Speed' value from the row
    speed: int = row['Speed']
    # accessing the 'Defense' value from the row
    defense: int = row['Defense']

    return sum([attack, speed, defense])
```

```python
pokemon['Score'] = pokemon.apply(score, axis=1)
pokemon
```
